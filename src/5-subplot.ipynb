### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for line plot
x = np.linspace(0, 10, 100)
y = x ** 2

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]


# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Line plot in the first subplot
axes[0, 0].plot(x, y, label='y = x^2', color='blue', linewidth=2)
axes[0, 0].set_title('Line Plot of y = x^2', fontsize=14)
axes[0, 0].set_xlabel('x values', fontsize=12)
axes[0, 0].set_ylabel('y values (x^2)', fontsize=12)
axes[0, 0].grid(True, linestyle='--', alpha=0.7)
axes[0, 0].legend(fontsize=10)

# Bar plot in the second subplot
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title('Bar Plot of Categories', fontsize=14)
axes[0, 1].set_xlabel('Categories', fontsize=12)
axes[0, 1].set_ylabel('Values', fontsize=12)

# Scatter plot in the third subplot
axes[1, 0].scatter(x_scatter, y_scatter, color='orange', label='Data Points')
for i, (x_val, y_val) in enumerate(zip(x_scatter, y_scatter)):
    axes[1, 0].annotate(f'({x_val}, {y_val})', (x_val, y_val), textcoords="offset points", xytext=(5, 5), fontsize=10)
axes[1, 0].set_title('Scatter Plot with Annotations', fontsize=14)
axes[1, 0].set_xlabel('x values', fontsize=12)
axes[1, 0].set_ylabel('y values', fontsize=12)
axes[1, 0].legend(fontsize=10)

# Pie chart in the fourth subplot
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink'])
axes[1, 1].set_title('Pie Chart of Programming Languages', fontsize=14)

# Adjust layout
plt.tight_layout()

# Show the 2x2 subplot layout
plt.show()
```
